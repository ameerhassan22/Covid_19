### clean and filter the data

```python
# find missing values
covid_19.isnull()
```

```python
## copy data set
train = covid_19.copy()
```

```python
## DATE_DIED = 9999-99-99 is alive now
def died(x):
    if x == '9999-99-99':
        return 0
    else:
        return 1


train['DEATH'] = train['DATE_DIED'].apply(died)
train.drop(columns=['DATE_DIED'], inplace=True)
```

```python
train
```

```python
corr = train.corr()
corr['DEATH']
```

```python
 train['Covid'] = [0 if i >= 4 else 1 for i in train.CLASIFFICATION_FINAL]
```

let's deal with our missing data (values of 97, 98 and 99):

We can notice that we have major missing values at the following features:

1- "ICU"

2- "INTUBED"

3- "PREGNANT

```python
## let's deal with our missing data in PREGNANT
train[train['SEX']==2]['PREGNANT']
```

```python
## we can replace all the corresponding values to the male section (97) with (2)
train['PREGNANT'].replace(97,2,inplace=True)
```

```python
## we can replace all the corresponding values to the male section (98) with (0)
train['PREGNANT'].replace(98,0,inplace=True)
```

 Unfortunatley we can't fill or replace the missing values of (98); because we can't tell which of these women were pregnant and which were not, and we can't also predict these values or fill them by random as it will affect our analysis which we are trying to build here based on only true and real life data

```python
train[train['PATIENT_TYPE']==1]['ICU']
```

```python
## we can replace all the corresponding values to the ICU section (97) with (2)
train['ICU'].replace(97,2,inplace=True)
```

```python
## we can replace all the corresponding values to the ICU section (99) with (0)
train['ICU'].replace(99,0,inplace=True)
```

```python
train[train['ICU']==2]['INTUBED']
```

Again and as predicted it's the same as the ICU case.
So we can replace all the values of (97) with (2); since obviously patients who have never been hospitalized couldn't possibly be connected to the ventilator.¶


```python
train['INTUBED'].replace (97, 2, inplace = True)
```
